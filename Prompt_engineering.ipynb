```python
# Real-world 2026 Example: Financial Risk Assessment using CoT
import json

# Simulating a call to a web search or financial API (PragyanAI context)
online_data = {
    "symbol": "TATAMOTORS",
    "current_price": 950.45,
    "news_sentiment": "Negative (Regulatory delay in EV subsidy)",
    "historical_volatility": "High",
    "market_trend": "Bearish"
}

# Few-Shot CoT Prompt Construction
system_prompt = "You are a Senior Risk Analyst. You process raw market data using expert Chain-of-Thought logic."

# We provide one 'Few-Shot' example that includes the 'Chain-of-Thought'
few_shot_example = """
Input: {"symbol": "RELIANCE", "sentiment": "Positive", "price": 2500}
Thought Process:
1. Analyze Sentiment: The positive sentiment suggests potential buy pressure.
2. Contextualize: Reliance is a heavyweight; its move affects the entire Nifty 50 index.
3. Risk Assessment: If price is near resistance, the positive sentiment might already be priced in.
Final Recommendation: HOLD until breakout above resistance.
"""

user_query = f"Process this online data: {json.dumps(online_data)}. Think through the regulatory impact."

if client:
    try:
        response = client.chat.completions.create(
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": few_shot_example},
                {"role": "user", "content": user_query}
            ],
            # 2026 reasoning models often have a 'reasoning_effort' parameter
            # model="llama-3.3-70b-versatile", # Using your current Llama setup
            model="llama-3.3-70b-versatile",
        )
        print("Model Reasoning and Output:")
        print(response.choices[0].message.content)
    except Exception as e:
        print(f"Integration Error: {e}")
```

```python
from llama_index.llms.groq import Groq
from llama_index.readers.web import BeautifulSoupWebReader
from pydantic import BaseModel, Field
from typing import List

# Initialize Groq
llm = Groq(model="llama-3.3-70b-versatile", api_key=userdata.get('prompt_engin'))
```

```python
class CollegePlacement(BaseModel):
    """Data model for university placement statistics."""
    university_name: str = Field(description="Name of the college or university")
    highest_package: str = Field(description="The highest salary package offered")
    average_package: str = Field(description="The average salary package")
    top_recruiters: List[str] = Field(description="List of major companies mentioned")

# Create a structured LLM
sllm = llm.as_structured_llm(CollegePlacement)
```

```python
# 1. Load the website data
reader = BeautifulSoupWebReader()
documents = reader.load_data(urls=["https://uvce.ac.in/placement_office.php","https://pes.edu/placements/"])

# 2. Extract structured data using a Prompt-based approach
# We provide the model with the web content and let it 'reason' out the fields.
prompt = f"Extract the placement details from the following web content:\n\n{documents[0].text}"
output = sllm.complete(prompt)

# This returns a Python object (Pydantic model)
placement_data = output.raw
print(f"University: {placement_data.university_name}")
print(f"Avg Package: {placement_data.average_package}")
print(f"Highest Package: {placement_data.highest_package}")
print(f"Top Recruiters: {placement_data.top_recruiters}")
```
